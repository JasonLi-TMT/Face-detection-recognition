```python
"""Generate detector, predictor, facerecognizer"""
#1 detector: detect face   input: dets eg:[[(115, 116) (223, 223)]]
#2 predictor: predict shape   input: eg: predictor(img, dets[0])
#3 facrec: generate feature array .   input: eg:

"""code eg:
img = io.imread(f)
dets = detector(img, 1)  # length of dets means number of face detected
for i in range(len(dets)):
    shape = predictor(img, dets[i])
    face_descriptor = facerec.compute_face_descriptor(img, shape)
    
"""

detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(predictor_path)
facerec = dlib.face_recognition_model_v1(face_rec_model_path)
```

```python
def img_encoder(file_path):
    """
    Function: input a file path, through a detector, predictor and encoding part
    
    Input: file_path is a path to the file
    Output: an array after decoding an image
    """
    
    img = io.imread(file_path)
    dets = detector(img, 1)
    shape = predictor(img, dets[0])
    face_descriptor = facerec.compute_face_descriptor(img, shape)
    return face_descriptor
```

```python
"""generate encoded image (arr) dictionary"""
def generate_face_db(doc_path):
    """
    Function: loop over the folder contains face and generate a face dictionary
    
    Input: doc_path to the project home folder
    
    Output: face_db: a dictionary, key is person's name, value is the encoding value (array) of a persion's face
    """
    
    face_db ={}
    for file_path,_,_ in os.walk(doc_path+"/img/"):
        # get the name of each face, eg: "jason li"
        name = os.path.split(file_path)[1]

        if name != "":
            print(name)

            # get all file path
            face_path_one_man = glob.glob(doc_path + '/img/' + name + '/*.jpg', recursive=True)

            
            face_encoding_arr = []
            for face_path in face_path_one_man:
                #generate empty list to save all array generated by img
                try:
                    face_encoding_arr.append([np.array(img_encoder(face_path)),face_path])
                except:
                    continue
            
            
            # assign the arr of face to the dictionary
            face_db[name] = np.array(face_encoding_arr)
    return face_db
```

```python
def compare_faces(face_encodings, face_to_compare,tolerance=0.4):
    """
    Given a list of face encodings, compare them to a known face encoding and get a euclidean distance
    for each comparison face. The distance tells you how similar the faces are.
    :param faces: List of face encodings to compare
    :param face_to_compare: A face encoding to compare against
    :return: A numpy ndarray with the distance for each face in the same order as the 'faces' array
    """
    if len(face_encodings) == 0:
        return np.empty((0))

    distance = np.linalg.norm(face_encodings - face_to_compare, axis=0)
    if distance<=tolerance:
        #print("Same person")
        return True
    else:
        #print("Not same person")
        return False
    

def compare_db(path,face_db):
    face_encodings = np.array(img_encoder(path))
    for name,value in face_db.items():
        for face_arr in face_db[name]:
            if compare_faces(face_encodings,face_arr[0]) == True:
                return name,face_arr[1]
    print("Not in database")
    return False
    
def visualize(path):
    """
    function: visualize the result, including bonding box and landmarks of a face
    
    Input: path: path to a file(jpg)
    Output: no output, just visualization
    
    """
    global win
    win = dlib.image_window()


    img = io.imread(path)

    win.clear_overlay()
    win.set_image(img)

    # Ask the detector to find the bounding boxes of each face. The 1 in the
    # second argument indicates that we should upsample the image 1 time. This
    # will make everything bigger and allow us to detect more faces.
    dets = detector(img, 1)
    print("Number of faces detected: {}".format(len(dets)))

    # Now process each face we found.
    for k, d in enumerate(dets):
        print("Detection {}: Left: {} Top: {} Right: {} Bottom: {}".format(
            k, d.left(), d.top(), d.right(), d.bottom()))
        # Get the landmarks/parts for the face in box d.
        shape = predictor(img, d)
        # Draw the face landmarks on the screen so we can see what face is currently being processed.
        win.clear_overlay()
        win.add_overlay(d)
        win.add_overlay(shape)


        dlib.hit_enter_to_continue()
```
